### Cálculo de raices Método de Newton vs Secante
---



- Vector xs
- Calcular vector f(xs) = vs


```python
xs = list(range(-10, 30))
```

```python
def f(x):
    _y = x**3 - 3 * x**2 + x - 1
    return _y
```

```python
ys = [f(x) for x in xs]
```

```python
import matplotlib.pyplot as plt

plt.plot(xs, ys)
plt.xlabel("xs")
plt.ylabel("ys")
plt.title("Gráfica de ys vs xs")
plt.grid(True)
plt.show()
```
